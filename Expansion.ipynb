```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# add stop words
stopwords = set(STOPWORDS)
stopwords.update(['Restaurants', "Restaurants,","Restaurants'", 'Food', "Food'","Food,","&","'Restaurants',","'Food,","'Restaurants,"])

# choose city
city_data = df[df['city'] == 'Newark']

# select city's categories text
categories_text = ' '.join(str(category).strip("',").strip("[]").strip("'") for category in city_data['categories'].str.split("', "))

# Create wordcloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(categories_text)

# plot
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from collections import Counter
# calculate words frequency
word_freq = Counter(categories_text.split(","))

# sort it
sorted_word_freq = dict(sorted(word_freq.items(), key=lambda x: x[1], reverse=True))
# show it
for word, freq in sorted_word_freq.items():
    if word not in stopwords:
        print(f"{word}: {freq}")
```
